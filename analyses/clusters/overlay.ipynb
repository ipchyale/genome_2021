```python
import sys
sys.path.append("/Users/damoncrockett/ivpy/src/")
from ivpy import *
from ivpy.glyph import draw_glyphs
from ivpy.extract import norm
from PIL import Image
import os
```

```python
import pandas as pd
```

```python
df = pd.read_csv("../../processing/genome/radar_impute.csv") # just the genome measurement data
```

```python
df[['t','c','g','r']] = norm(df[['thickness','LAB_B_dmin','g60','logroughness']])
```

```python
df.g = [1-item for item in df.g] # because gloss is inverse
```

```python
df.catalog = df.catalog.apply(str)
```

```python
aes = {
    "top":"c",
    "left":"t",
    "right":"g",
    "bottom":"r",
    "basename":"catalog"
}
```

```python
GDIR = "/Users/damoncrockett/genome_2021/analyses/clusters/glyphs/"
```

```python
try:
    os.mkdir(GDIR)
except:
    pass
```

```python
df['gpath'] = draw_glyphs(df,aes,GDIR,gridlines=False,mat=False,radii=False,side=800,alpha=0.01)
```

```python
from ivpy.cluster import cluster
```

```python
df['cluster'] = cluster(df[['t','c','g','r']],method='hierarchical',k=5)
```

```python
canvas = Image.new('RGB',(800,800),"#212121")
for gpath in df.gpath[df.cluster==0]:
    im = Image.open(gpath)
    canvas.paste(im,(0,0),im)
canvas
```
